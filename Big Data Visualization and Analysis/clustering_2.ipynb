## One successful example for k-means, hierarchical clustering and dbscan

```python
from sklearn.cluster import DBSCAN
from sklearn.cluster import KMeans
from sklearn.cluster import AgglomerativeClustering

from sklearn.metrics.cluster import adjusted_rand_score
import pandas as pd
import matplotlib.pyplot as plt
```

```python
sample_df = pd.read_csv("C:/Users/username/Downloads/fourty.csv")

training_points = sample_df[["col1", "col2"]]
training_labels = sample_df["target"]

kmeans = KMeans(n_clusters=40).fit(training_points)

arc = adjusted_rand_score(training_labels, kmeans.labels_)

plt.scatter(training_points["col1"], training_points["col2"], c=kmeans.labels_, 
	cmap='rainbow')

plt.show()

print(arc)
```

```python
sample_df = pd.read_csv("C:/Users/username/Downloads/fourty.csv")

training_points = sample_df[["col1", "col2"]]
training_labels = sample_df["target"]

agglo = AgglomerativeClustering(n_clusters=40, linkage="single").fit(training_points)

arc = adjusted_rand_score(training_labels, agglo.labels_)

plt.scatter(training_points["col1"], training_points["col2"], c=agglo.labels_, 
	cmap='rainbow')

plt.show()

print(arc)
```

```python
sample_df = pd.read_csv("C:/Users/username/Downloads/fourty.csv")

training_points = sample_df[["col1", "col2"]]
training_labels = sample_df["target"]

max = 0.0
a = 0.1
b=1

for a in range(1, 100, 1):
    for b in range(1, 20, 1):
        dbscan = DBSCAN(eps = a*0.01, min_samples = b).fit(training_points)
        arc = adjusted_rand_score (training_labels, dbscan. labels_)
        if (max < arc):
            max = arc
            result_a = a
            result_b = b

dbscan = DBSCAN(eps=result_a * 0.01, min_samples=result_b).fit(training_points)

plt.scatter(training_points["col1"], training_points["col2"], c=dbscan.labels_, 
	cmap='rainbow')

plt.show()

print(result_a * 0.01)
print(result_b)
print("arc :",end="")
print(max)
```
