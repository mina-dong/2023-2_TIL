## One successful example for k-means, hierarchical clustering and dbscan

```python
from sklearn.cluster import KMeans
import pandas as pd
from sklearn.metrics.cluster import adjusted_rand_score
import matplotlib.pyplot as plt

sample_df = pd.read_csv("C:/Users/username/Downloads/insect.csv")

training_points = sample_df[["col1", "col2","col3"]]
training_labels = sample_df["target"]

kmeans = KMeans(n_clusters=2).fit(training_points)

arc = adjusted_rand_score(training_labels, kmeans.labels_)

plt.scatter(training_points["col1"], training_points["col2"], training_points["col3"], c=kmeans.labels_, 
	cmap='rainbow')

plt.show()

print(arc)
```

```python
from sklearn.cluster import DBSCAN
import pandas as pd
import matplotlib.pyplot as plt

sample_df = pd.read_csv("C:/Users/username/Downloads/insect.csv")

training_points = sample_df[["col1", "col2","col3"]]
training_labels = sample_df["target"]

max = 0.0
a = 0.1
b=1

for a in range(1, 100, 1):
    for b in range(1, 20, 1):
        dbscan = DBSCAN(eps = a*0.01, min_samples = b).fit(training_points)
        arc = adjusted_rand_score (training_labels, dbscan. labels_)
        if (max < arc):
            max = arc
            result_a = a
            result_b = b

dbscan = DBSCAN(eps=result_a * 0.01, min_samples=result_b).fit(training_points)

plt.scatter(training_points["col1"], training_points["col2"], training_points["col3"], c=dbscan.labels_, 
	cmap='rainbow')

plt.show()

print(result_a * 0.01)
print(result_b)
print("arc :",end="")
print(max)
```
