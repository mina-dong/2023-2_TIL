### Data Cleaning

```python
# Data cleaning
df.drop(columns = 'Other_Sales', inplace = True)
df.head()
```

### Data Transformation

```python
# Data Transformation
df.columns

df.head()
```

```python
# Data Transformation
df.rename(columns={
        'Name': 'GameName', 
}, inplace=True)

df.columns
```

### Data Normalization

```python
df.describe()
```

```python
# Data Normalization - Min-max normalization
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

df[['Global_Sales']] = scaler.fit_transform(df[['Global_Sales']])
df[['NA_Sales']] = scaler.fit_transform(df[['NA_Sales']])
df[['EU_Sales']] = scaler.fit_transform(df[['EU_Sales']])
df[['JP_Sales']] = scaler.fit_transform(df[['JP_Sales']])

df.describe()
```
