# Use on of the dataframes provided by Seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.get_dataset_names()
```

```python
data = sns.load_dataset('taxis')

data.head()
```

# Apply basic charts : Bar and scatter charts

```python
#Bar Charts
sns.countplot(x='color', 
	data=data, 
	hue="pickup_borough", 
	palette="Greens")

plt.title("Pickup borough by Texis Color")
plt.show()
```

택시 색상을 기준으로 개수를 세움. 노란택시가 맨하탄지구에서 압도적으로 택시를 타고 있음을 알수있음 = 노란색 택시가 맨하탄지구에서 초록 택시보다 더 많이 돌아다님.

```python
#Scatter Charts
sns.scatterplot(x='distance', y = 'fare', data = data, hue = 'payment')

plt.title("Fare prices by Deck")
plt.show()
```

```python
거리별 요금에 대한 것이다. 거리와 요금은 상관관계까 있으며  
비교적 긴거리와 짧은거리는 현금으로 지불했으며 
중간길이거리에서는 주로 카드로 지불함.
```

# Apply statistical charts : Boxplot, violin and joint plots

```python
# Boxplot
sns.boxplot(x = 'color', y='fare', data=data, palette='Paired')

plt.show()
```

```python
노란색 택시와 초록색 택시를 비교해봤을 때 비슷해보이나, 노란택시가 훨씬 더 많은 요금을 지불된 것이 많다고 확인할 수 있음.
```

```python
# Violinplot
sns.violinplot(x = 'color', y='fare', data=data, palette='Paired')

plt.show()
```

```python
폭의 너비를 봤을 때 녹색보다 노란색택시가 더 많은 사람이 이용하고 있음을 확인할 수 있음.
```

```python
# jointplot
sns.jointplot(x = 'fare', y='tip', hue='dropoff_borough', data=data)

plt.show()
```

```python
요금이 낮는 안낮든 팁을 안내는 손님은 여러 자치구에 있으나, 맨하탄에 내린 사람들은 대부분 팁을 잘낸다
```

# Study at least two more by yourself and apply it to the dataset

```python
#stripplot : 연속형 변수와 범주형 변수
sns.stripplot(x="passengers", y="total", data=data)
plt.show()
```

```python
탑승객별 요금을 확인할 수 있다. 비교적 적은 인원일때 더 많은 요금을 내고, 사람이 많아질수록 비교적 적은 요금을 지불한다.
```

```python
#missing value
sns.heatmap(data.isnull(), cbar=False)

plt.show()
```

```python
결측치를 보여주는 그래프이다. 요금과 같은 데이터는 기록이 잘 되어있고, 지역과 관련된 데이터는 요금과 관련된 데이터에 비해 덜 기록 되었다.
```
